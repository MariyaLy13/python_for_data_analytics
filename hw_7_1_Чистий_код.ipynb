````python
input_list = [75, 80, 85, 90, 95]
````

````python
def calculate_variance(nl):
    sum_list = 0
    for n in nl:
        sum_list += n  # Use shorthand for addition
    mean = sum_list / len(nl)

    sum_squares = 0
    for n in nl:
        sum_squares += n ** 2  # Use shorthand for addition

    mean_squares = sum_squares / len(nl)
    variance = mean_squares - mean ** 2  # Calculate variance
    return variance  # Return the calculated variance
````

````python
print(calculate_variance(input_list))
````
