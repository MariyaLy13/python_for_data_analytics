### Lifestyle Factors impact

**6. What is the BMI of patients with a positive diagnosis in compare to a negative diagnosis?**

```python
# Group by Diagnosis and calculate BMI statistics (min, max, mean, median)
bmi_stats = df.groupby('Diagnosis')['BMI'].agg(['min', 'max', 'mean', 'median'])

print('BMI Statistics:')
print(bmi_stats)
```

- BMI is under 18.5 - you are very underweight and possibly malnourished.
- BMI is between 18.5 and 24.9 – This is described as the 'healthy range'. 
- BMI is between 25 and 29.9 – This is described as overweight. 
- BMI is 30 or over – This is described as obesity.

```python
# Define BMI ranges
BMI_bins = [0, 18.5, 25, 30, 40]
BMI_labels = ['BMI < 18.5 - underweight', 'BMI 18.5-24.9 - healthy range', 'BMI 25-29.9 - overweight', 'BMI > 30 - obesity']

# Assign BMI categories to the DataFrame
df['BMI_range'] = pd.cut(df['BMI'], bins=BMI_bins, labels=BMI_labels, right=True)

# Calculate the proportions of each BMI category by Diagnosis
BMI_range_proportions = df.groupby(['Diagnosis', 'BMI_range'], observed=False).size() / df.groupby('Diagnosis', observed=False).size()

# Reshape the result for easier visualization
BMI_range_df = BMI_range_proportions.unstack(fill_value=0)

print('BMI Range Proportions by Diagnosis:')
print(BMI_range_df)
```

```python
# Plotting the BMI categories by Diagnosis
BMI_range_df.plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('BMI Categories by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Proportion')
plt.xticks([0, 1], ['Negative Diagnosis (0)', 'Positive Diagnosis (1)'], rotation=0)
plt.legend(title='BMI Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Medical History impact

- **12. What is the % of patients with a positive diagnosis have Diabetes?**
- **13. What is the % of patients with a positive diagnosis have FamilyHistoryAlzheimers?**
- **14. What is the % of patients with a positive diagnosis have Cardiovascular Disease?**
- **15. What is the % of patients with a positive diagnosis have HeadInjury?**
- **16. What is the % of patients with a positive diagnosis have Depression?**
- **17. What is the % of patients with a positive diagnosis have Hypertension?**

```python
# List of columns related to medical history
medical_history_columns = ['Diabetes', 'FamilyHistoryAlzheimers', 'CardiovascularDisease', 
                           'HeadInjury', 'Depression', 'Hypertension']

# Initialize an empty list to store the results
proportion_results = []

# Loop through each column to calculate the proportions
for column in medical_history_columns:
    # Group by Diagnosis and the current medical history column, calculate counts
    counts = df.groupby(['Diagnosis', column]).size().unstack(fill_value=0)
    
    # Calculate proportions by dividing each count by the total for the corresponding Diagnosis
    proportions = counts.div(counts.sum(axis=1), axis=0)
    
    # Add the results to the list
    proportion_results.append(proportions)

# Concatenate all results into one DataFrame for better readability
final_results = pd.concat(proportion_results, keys=medical_history_columns, axis=1)

# Display the final results
print("Proportions of Medical History Impact by Diagnosis:")
print(final_results)
```

```python
# Plotting the results
plt.figure(figsize=(14, 12))

# Create a bar plot for each medical history condition
for idx, column in enumerate(medical_history_columns):
    data = final_results[column]
    
    # Plot each condition separately
    ax = plt.subplot(2, 3, idx + 1)  # 2 rows, 3 columns
    bars = data.plot(kind='bar', ax=ax, color=['lightblue', 'lightgreen'])
    
    ax.set_title(f'{column} Proportions by Diagnosis')
    ax.set_ylabel('Proportion')
    ax.set_xlabel('Diagnosis')
    ax.set_xticklabels(['Negative Diagnosis (0)', 'Positive Diagnosis (1)'], rotation=0)
    ax.legend(['No', 'Yes'], title=column)
    
    # Add percentage labels on the bars
    for bar in bars.patches:
        height = bar.get_height()  # Get the height of each bar
        percentage = f'{height * 100:.1f}%'  # Convert to percentage and format to 1 decimal place
        ax.text(bar.get_x() + bar.get_width() / 2, height + 0.02, percentage, 
                ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()  # To prevent overlap of subplots
plt.show()
```

```python
# Define the bins and labels
Systolic_BP_bins = [90, 120, 150, 180]
Systolic_BP_labels = ['90 - 120 mmHg', '120 - 150 mmHg', '150 - 180 mmHg']

Diastolic_BP_bins = [60, 80, 100, 120]
Diastolic_BP_labels = ['60 - 80 mmHg', '80 - 100 mmHg', '100 - 120 mmHg']

Cholesterol_Total_bins = [150, 200, 250, 300]
Cholesterol_Total_labels = ['150 - 200 mg/dL', '200 - 250 mg/dL', '250 - 300 mg/dL']

Cholesterol_LDL_bins = [50, 100, 150, 200]
Cholesterol_LDL_labels = ['50 - 100 mg/dL', '100 - 150 mg/dL', '150 - 200 mg/dL']

Cholesterol_HDL_bins = [20, 40, 60, 80, 100]
Cholesterol_HDL_labels = ['20 - 40 mg/dL', '40 - 60 mg/dL', '60 - 80 mg/dL', '80 - 100 mg/dL']

Triglycerides_bins = [50, 150, 250, 350, 400]
Triglycerides_labels = ['50 - 150 mg/dL', '150 - 250 mg/dL', '250 - 350 mg/dL', '350 - 400 mg/dL']

# Create the range for each column in the DataFrame
df['SystolicBP_range'] = pd.cut(df['SystolicBP'], bins=Systolic_BP_bins, labels=Systolic_BP_labels, right=False)
df['DiastolicBP_range'] = pd.cut(df['DiastolicBP'], bins=Diastolic_BP_bins, labels=Diastolic_BP_labels, right=False)

df['CholesterolTotal_range'] = pd.cut(df['CholesterolTotal'], bins=Cholesterol_Total_bins, labels=Cholesterol_Total_labels, right=False)
df['CholesterolLDL_range'] = pd.cut(df['CholesterolLDL'], bins=Cholesterol_LDL_bins, labels=Cholesterol_LDL_labels, right=False)
df['CholesterolHDL_range'] = pd.cut(df['CholesterolHDL'], bins=Cholesterol_HDL_bins, labels=Cholesterol_HDL_labels, right=False)

df['CholesterolTriglycerides_range'] = pd.cut(df['CholesterolTriglycerides'], bins=Triglycerides_bins, labels=Triglycerides_labels, right=False)
```

```python
# List of columns related to Clinical Measurements
clinical_measurements_columns = ['SystolicBP_range', 'DiastolicBP_range', 'CholesterolTotal_range', 
                           'CholesterolLDL_range', 'CholesterolHDL_range', 'CholesterolTriglycerides_range']

# Initialize an empty list to store the results
clinical_proportion_results = []

# Loop through each column to calculate the proportions
for column in clinical_measurements_columns:
    # Group by Diagnosis and the current clinical measurement column, calculate counts
    counts = df.groupby(['Diagnosis', column], observed=False).size().unstack(fill_value=0)
    
    # Calculate proportions by dividing each count by the total for the corresponding Diagnosis
    proportions = counts.div(counts.sum(axis=1), axis=0)
    
    # Add the results to the list
    clinical_proportion_results.append(proportions)

# Concatenate all results into one DataFrame for better readability
clinical_measurements_final_results = pd.concat(clinical_proportion_results, keys=clinical_measurements_columns, axis=1)

# Display the final results
print("Proportions of Clinical Measurements Impact by Diagnosis:")
print(clinical_measurements_final_results)
```

### Chi-Square Test for Independence:
The Chi-square test helps to assess whether there is a significant relationship between two categorical variables (e.g., the presence of comorbid conditions and Alzheimer's diagnosis).
Steps for Chi-Square Test:
- Null Hypothesis (H0): There is no relationship between comorbid conditions and Alzheimer's diagnosis.
- Alternative Hypothesis (H1): There is a significant relationship between comorbid conditions and Alzheimer's diagnosis.

In this case, you’ll test whether each individual comorbid condition (e.g., Cardiovascular Disease, Diabetes, Depression, Hypertension, Head Injury) is independent of the Alzheimer’s diagnosis (Diagnosis column).

```python
# List of comorbid conditions columns
comorbid_columns = ['FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury', 'Hypertension']

# Perform chi-square test for each comorbid condition vs Alzheimer's diagnosis
for column in comorbid_columns:
    contingency_table = pd.crosstab(df[column], df['Diagnosis'])
    chi2, p_value, _, _ = stats.chi2_contingency(contingency_table)
    
    # Print the result
    print(f"Chi-square test for {column} and Alzheimer's diagnosis:")
    print(f"Chi2 Statistic: {chi2}, p-value: {p_value}\n")
    
    if p_value < 0.05:
        print(f"Result: Significant relationship between {column} and Alzheimer's diagnosis.\n")
    else:
        print(f"Result: No significant relationship between {column} and Alzheimer's diagnosis.\n")
```

### Logistic Regression:

If you want to model the relationship between multiple comorbid conditions and the likelihood of Alzheimer's diagnosis (i.e., predict the probability of Alzheimer's diagnosis based on comorbidities), you can use logistic regression.

Logistic regression is appropriate for situations where the dependent variable (target) is categorical, such as in the case of Alzheimer's diagnosis (Diagnosis), which is a binary variable (1 = Alzheimer's, 0 = No Alzheimer's).

Steps for Logistic Regression:
- Dependent variable (y): Diagnosis (Alzheimer’s diagnosis).
- Independent variables (X): Comorbid conditions (e.g., CardiovascularDisease, Diabetes, Depression, HeadInjury, Hypertension).

```python
import statsmodels.api as sm
```

```python
# Select relevant columns for logistic regression
x_regression = df[['FamilyHistoryAlzheimers', 'CardiovascularDisease', 'Diabetes', 'Depression', 'HeadInjury', 'Hypertension']]
y_regression = df['Diagnosis']

# Add a constant to the independent variables matrix (for the intercept term)
X_regression = sm.add_constant(x_regression)

# Fit the logistic regression model
model_regression = sm.Logit(y_regression, X_regression)
result_regression = model_regression.fit()

# Print the summary of the model
print(result_regression.summary())
```

```python
# Create the logistic regression model
x_unhealthy = df[['Smoking', 'AlcoholConsumption', 'PhysicalActivity', 
                  'DietQuality', 'SleepQuality']]
y_unhealthy = df['Diagnosis']

# Add a constant (intercept) to the model
X_unhealthy = sm.add_constant(x_unhealthy)

# Fit the logistic regression model
model_unhealthy = sm.Logit(y_unhealthy, X_unhealthy)
result_unhealthy = model_unhealthy.fit()

# Display the summary of the model
print(result_unhealthy.summary())
```
